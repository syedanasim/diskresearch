```python
def sdensity_rauch(arg): 
    sdensitycal= 2*(1e-3)*((arg)**(-3/4))*(msun*1000)*((rsun*100)**-2)
    return sdensitycal

def sdensity_plot(): 
    #creating a list of values for (density) disk density
    arg_rauch=[1e1,1e2,1e3,1e4,1e5]
    acm_rauch=[1e1*100*Rg,1e2*100*Rg,1e3*100*Rg,1e4*100*Rg,1e5*100*Rg]
    densityden_rauch=[]
    for i in range(len(arg_rauch)):
        densityden_rauch.append(sdensity_rauch(arg_rauch[i]))
    
    fig=plt.figure(figsize=(10,8))
    ax=plt.subplot(111)
    #ax.plot(acm_rauch,densityden_rauch,lw=2,label='Rauch',c='k')
    ax.plot(arg_rauch,densityden_rauch,lw=2,label='Rauch',c='k')
    ax.set_title('Rauch Surface Density')
    ax.set_xlabel('$a$ [$Rg$]')  #cm
    ax.set_ylabel('$σ$ [$g/cm^3$]') #$density$     [$msun * rsun{^-2}$]
    ax.legend(fontsize=8, framealpha=1, edgecolor= 'k', handlelength=2.8, 
              ncol=2, columnspacing=0.5, loc='center left', bbox_to_anchor=(0,0.05))
    plt.xscale('log')
    plt.yscale('log')
    plt.savefig('plots/propertyplots/RauchSurfaceDensity.pdf')
```

```python
import scipy as sp
import scipy.interpolate

def log_interp1d(xx, yy, kind='linear'):
    logx = np.log10(xx)
    logy = np.log10(yy)
    lin_interp = sp.interpolate.interp1d(logx, logy, kind=kind)
    log_interp = lambda zz: np.power(10.0, lin_interp(np.log10(zz)))
    return log_interp
```

```python
def heightplot(): 
    logPlot=plt.gca()
    logPlot.loglog(radiusSG/Rg,heightSG,lw=2,label='SG',c='b')
    logPlot.loglog(radiusTQM/Rg,heightTQM,lw=2,label='TQM',c='r')
    plt.title('Height')
    logPlot.set_xlabel('$a$ [$R_g$]')
    logPlot.set_ylabel('$H$ [$m$]')
    logPlot.legend()
    ax = plt.subplot(111)
    ax.legend(loc='center left', bbox_to_anchor=(0.8,0.1))
def surfacedensityplot(): 
    logPlot=plt.gca()
    logPlot.loglog(radiusSG/Rg,surfacedensitySG,lw=2,label='SG',c='b')
    logPlot.loglog(radiusTQM/Rg,surfacedensityTQM,lw=2,label='TQM',c='r')
    plt.title('Surface Density')
    logPlot.set_xlabel('$a$ [$R_g$]')
    logPlot.set_ylabel('$surfacedensity$ [$g/cm^2$]')
    logPlot.legend()
    ax = plt.subplot(111)
    ax.legend(loc='center left', bbox_to_anchor=(0.8,0.9))
```
